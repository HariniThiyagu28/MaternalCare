```python
# Home page
%%writefile pages/1_Home.py
import streamlit as st
from datetime import datetime
import random
import json
import os


st.set_page_config(page_title="SmartMaternalCare", layout="wide")


# Custom styles for column backgrounds and spacing
st.markdown("""
    <style>
    /* Full-page gradient */
    .stApp {
        background: linear-gradient(to bottom, #fcb0b3 0%, #d8b4e2 50%, #add8e6 100%);
        min-height: 100vh;
        color: #000;
        font-size: 18px;
    }
    .main, .block-container {
        background-color: transparent !important;
    }
    /* Section backgrounds: off-white cards for these 3 sections */
    .section-col-offwhite {
        background: #f8f8f8 !important;  /* subtle off-white */
        padding: 15px 20px;
        border-radius: 8px;
        box-shadow: 1px 1px 5px rgba(0,0,0,0.05);
        margin-bottom: 15px;
    }
    </style>
""", unsafe_allow_html=True)


def calculate_weeks_pregnant(edd):
    if edd is None:
        return None
    today = datetime.today().date()
    delta = edd - today
    weeks_left = delta.days / 7
    weeks_pregnant = max(0, min(40, int(40 - weeks_left)))
    return weeks_pregnant


def baby_size_info(week):
    sizes = [
        (1, "🌱", "a tiny seed"),
        (4, "🌾", "fresh sprouting wheat"),
        (8, "🍐", "a pear"),
        (12, "🍅", "a tomato"),
        (16, "🥭", "a ripe mango"),
        (20, "🍌", "a banana"),
        (24, "🍍", "a pineapple"),
        (28, "🥥", "a coconut"),
        (32, "🎱", "a billiard ball"),
        (36, "🍉", "a watermelon"),
        (40, "🎂", "a cake")
    ]
    emoji, desc = "🤰", "your little miracle"
    if week is None or week == 0:
        return emoji, "your little miracle"
    for w, emj, phrase in reversed(sizes):
        if week >= w:
            return emj, phrase
    return emoji, desc


def load_todo(username):
    todo_path = f"{username}_todo.json"
    if os.path.exists(todo_path):
        try:
            with open(todo_path, "r") as f:
                return json.load(f)
        except Exception:
            # Corrupted file fallback
            return {"Take Folic Acid": "Pending", "Take Iron Tablet": "Pending", "Calcium Supplement": "Pending"}
    else:
        return {"Take Folic Acid": "Pending", "Take Iron Tablet": "Pending", "Calcium Supplement": "Pending"}


def save_todo(username, todo_list):
    todo_path = f"{username}_todo.json"
    with open(todo_path, "w") as f:
        json.dump(todo_list, f)


def main():
    st.title("🏠 SmartMaternalCare - Your Pregnancy Dashboard")
    st.markdown(" dear mom! Let's cherish every moment of your pregnancy journey with love and care 💕")

    # Initialize session state variables
    if "mom_name" not in st.session_state:
        st.session_state.mom_name = ""
    if "edd" not in st.session_state:
        st.session_state.edd = None
    if "new_med_text" not in st.session_state:
        st.session_state.new_med_text = ""
    if "new_vaccine_select" not in st.session_state:
        st.session_state.new_vaccine_select = "--Select Vaccine--"
    if "new_vaccine_week" not in st.session_state:
        st.session_state.new_vaccine_week = 1
    if "activities_done" not in st.session_state:
        st.session_state.activities_done = {}
    if "vaccinations" not in st.session_state:
        st.session_state.vaccinations = [
            {"name": "Tetanus", "due_week": 16, "status": "Pending"},
            {"name": "Influenza", "due_week": 32, "status": "Pending"},
        ]
    if "reset_vaccine_select" not in st.session_state:
        st.session_state.reset_vaccine_select = False

    # Flag for clearing medicine input after add
    if "med_added" not in st.session_state:
        st.session_state.med_added = False

    known_vaccines = ["Tetanus", "Influenza", "Hepatitis B", "Chickenpox"]

    # Reset vaccine selectbox value before widget if needed
    if st.session_state.reset_vaccine_select:
        st.session_state.new_vaccine_select = "--Select Vaccine--"
        st.session_state.reset_vaccine_select = False

    # Row 1: Name and Due Date inputs (no preset due date)
    col1, col2 = st.columns([1, 1], gap="small")
    with col1:
        mom_name = st.text_input("What's your name, lovely mom?", value=st.session_state.mom_name)
    with col2:
        edd = st.date_input(
            "Please enter your estimated due date (EDD) 🎉",
            value=st.session_state.edd if st.session_state.edd else datetime.today().date(),
            min_value=datetime.today().date()
        )

    st.session_state.mom_name = mom_name
    st.session_state.edd = edd

    if not mom_name:
        st.info("Please enter your name to personalize your journey.")
        return
    if not edd:
        st.info("Please enter your estimated due date (EDD).")
        return

    weeks_pregnant = calculate_weeks_pregnant(edd)

    if weeks_pregnant == 0:
        baby_emoji = "🎈"
        baby_comment = "your little miracle"
    else:
        baby_emoji, baby_comment = baby_size_info(weeks_pregnant)

    # Row 2: Baby size + days left
    col_baby, col_countdown = st.columns([1, 1], gap="small")
    with col_baby:
        st.markdown(f"### Your baby is about the size of a {baby_comment} {baby_emoji}")
        if weeks_pregnant is not None:
            st.write(f"**{weeks_pregnant} weeks pregnant**")

    with col_countdown:
        days_left = (edd - datetime.today().date()).days
        st.markdown(f"### ⏳ {days_left} days left, mama!")
        st.progress(min(max(weeks_pregnant / 40 if weeks_pregnant else 0, 0), 1))

    st.markdown("---")

    # Row 3: Three equal width columns with larger gap
    col_med, col_vacc, col_activity = st.columns(3, gap="large")

    # Helper functions for button callbacks
    def mark_medicine_taken(task):
        todo_list = load_todo(st.session_state.mom_name)
        todo_list[task] = "Taken"
        save_todo(st.session_state.mom_name, todo_list)
        st.rerun()

    def remove_medicine(task):
        todo_list = load_todo(st.session_state.mom_name)
        if task in todo_list:
            del todo_list[task]
            save_todo(st.session_state.mom_name, todo_list)
        st.rerun()

    def mark_vaccine_taken(index):
        st.session_state.vaccinations[index]["status"] = "Taken"
        st.rerun()

    # Medicine Reminder Section with off-white background
    with col_med:
        st.markdown('<div class="section-col-offwhite">', unsafe_allow_html=True)
        st.header("📝 Medicine Reminder")

        # Load todo fresh every time
        todo_list = load_todo(st.session_state.mom_name)

        # Reset the new_med_text input if medicine was just added
        if st.session_state.med_added:
            st.session_state.new_med_text = ""
            st.session_state.med_added = False

        for task, status in todo_list.items():
            cols = st.columns([3, 1, 1], gap="small")
            with cols[0]:
                st.write(f"🔸 {task}")
            with cols[1]:
                if status == "Pending":
                    if st.button("Take", key=f"take_{task}", on_click=mark_medicine_taken, args=(task,)):
                        pass
                else:
                    st.markdown("✅ Taken")
            with cols[2]:
                if st.button("Rem", key=f"remove_{task}", on_click=remove_medicine, args=(task,)):
                    pass

        new_med_text = st.text_input("Add new medicine reminder:", value=st.session_state.new_med_text, key="new_med_text")

        def add_medicine():
            med_to_add = st.session_state.new_med_text.strip()
            if med_to_add and med_to_add not in todo_list:
                todo_list[med_to_add] = "Pending"
                save_todo(st.session_state.mom_name, todo_list)
                st.session_state.med_added = True
                st.rerun()

        if st.button("Add Medicine"):
            add_medicine()

        st.markdown("</div>", unsafe_allow_html=True)

    # Vaccination Schedule Section with off-white background
    with col_vacc:
        st.markdown('<div class="section-col-offwhite">', unsafe_allow_html=True)
        st.header("💉 Vaccination Schedule")

        for i, v in enumerate(sorted(st.session_state.vaccinations, key=lambda x: x["due_week"])):
            cols_list = st.columns([2, 1], gap="small")
            cols_list[0].write(f"- {v['name']} (Due by Week {v['due_week']})")
            if v.get("status", "Pending") == "Pending":
                if cols_list[1].button("Take", key=f"take_vaccine_{i}", on_click=mark_vaccine_taken, args=(i,)):
                    pass
            else:
                cols_list[1].markdown("✅ Taken")

        new_vaccine_name = st.selectbox(
            "Select vaccine:",
            options=["--Select Vaccine--"] + known_vaccines,
            index=(known_vaccines.index(st.session_state.new_vaccine_select) + 1) if st.session_state.new_vaccine_select in known_vaccines else 0,
            key="new_vaccine_select"
        )
        new_vaccine_week = st.number_input(
            "Due week",
            min_value=1,
            max_value=40,
            value=st.session_state.new_vaccine_week,
            key="new_vaccine_week"
        )

        def add_vaccine():
            if st.session_state.new_vaccine_select != "--Select Vaccine--":
                name = st.session_state.new_vaccine_select
                week = st.session_state.new_vaccine_week
                st.session_state.vaccinations.append({"name": name, "due_week": week, "status": "Pending"})
                st.session_state.new_vaccine_week = 1
                st.session_state.reset_vaccine_select = True
                st.rerun()

        if st.button("Add Vaccine"):
            add_vaccine()

        st.markdown("</div>", unsafe_allow_html=True)

    # Activity Done Today Section with off-white background
    with col_activity:
        st.markdown('<div class="section-col-offwhite">', unsafe_allow_html=True)
        st.header("🌈 Activity Done Today")

        activities = {
            "Walking": st.checkbox("Walking", key="act_walking"),
            "Gentle Exercise": st.checkbox("Gentle Exercise", key="act_exercise"),
            "Meditation": st.checkbox("Meditation", key="act_meditation"),
            "Healthy Eating": st.checkbox("Healthy Eating", key="act_eating"),
            "Resting": st.checkbox("Resting", key="act_resting"),
        }

        if st.button("Save Activity"):
            done_activities = [act for act, done in activities.items() if done]
            if done_activities:
                st.success(f"Great job, mama! You completed: {', '.join(done_activities)} 💖")
            else:
                st.info("No activity selected yet. Try to do something nice for yourself today!")

        st.markdown("</div>", unsafe_allow_html=True)

    st.markdown("---")

    # Daily Tip below the 3-column row
    daily_tips = [
        "Stay hydrated by drinking at least 8 glasses of water daily. It helps digestion, circulation, and prevents fatigue.",
        "Include iron-rich foods like spinach, lentils, and red meat to support your baby’s growth and prevent maternal anemia.",
        "Engage in gentle exercises like prenatal yoga or walking to ease discomfort, boost mood, and prepare your body for labor.",
        "Always listen to your body—if you're tired, take time to rest. Proper sleep is essential for you and your baby.",
        "Practice deep breathing or mindfulness daily. It reduces stress, lowers blood pressure, and nurtures a calm, positive pregnancy journey.",
        "Take short walks throughout the day. It improves circulation, reduces swelling, and keeps your muscles active and flexible during pregnancy.",
        "Add colorful fruits and vegetables to each meal. They provide essential vitamins and fiber for both your health and your baby’s.",
    ]

    tip = random.choice(daily_tips)

    st.markdown("### 💡 Daily Tip")
    st.markdown(f"_{tip}_")

    st.markdown("**Thank you for allowing SmartMaternalCare to accompany you on this beautiful journey of motherhood! 💖 Remember to cherish each precious moment and keep an eye on your little one’s amazing growth and development in the tabs below. 🌸👶**")


if __name__ == "__main__":
    main()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

# Load dataset
df = pd.read_csv("Maternal_Cleaned.csv", skipinitialspace=True)

# Features and Target
X = df[['Age', 'SystolicBP', 'DiastolicBP', 'BS', 'BodyTemp', 'HeartRate']]
y = df['RiskLevel']

# Encode target labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Function to train and evaluate SVM with different kernels
def train_evaluate_svm(kernel_type):
    svm_model = SVC(kernel=kernel_type, probability=True)
    svm_model.fit(X_train_scaled, y_train)
    svm_preds = svm_model.predict(X_test_scaled)
    svm_acc = accuracy_score(y_test, svm_preds)
    print(f"✅ SVM Accuracy ({kernel_type} kernel):", svm_acc)
    print(f"\nSVM ({kernel_type} kernel) Report:\n", classification_report(y_test, svm_preds, target_names=le.classes_))
    return svm_model

# Try 'rbf', 'poly', and 'sigmoid' kernels
rbf_model = train_evaluate_svm('rbf')
poly_model = train_evaluate_svm('poly')
sigmoid_model = train_evaluate_svm('sigmoid')
```

```python
import os
import random
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader, WeightedRandomSampler
from collections import Counter
from tqdm import tqdm
from torch.amp import autocast, GradScaler

# Set seeds and device
torch.manual_seed(42)
random.seed(42)
np.random.seed(42)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Dataset paths
base_dir = "/content/drive/MyDrive/mini_pro"
train_dir = os.path.join(base_dir, 'train')
val_dir = os.path.join(base_dir, 'validation')
test_dir = os.path.join(base_dir, 'test')

# Hyperparameters
img_size = 224
batch_size = 8
num_classes = 3
num_epochs = 1000
patience = 10

# Data augmentations
train_transforms = transforms.Compose([
    transforms.Resize((img_size, img_size)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(15),
    transforms.RandomAffine(degrees=0, translate=(0.1, 0.3), shear=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225]),
])

val_test_transforms = transforms.Compose([
    transforms.Resize((img_size, img_size)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225]),
])

# Load datasets
train_dataset = datasets.ImageFolder(train_dir, transform=train_transforms)
val_dataset = datasets.ImageFolder(val_dir, transform=val_test_transforms)
test_dataset = datasets.ImageFolder(test_dir, transform=val_test_transforms)

# Compute class weights and create sampler to handle imbalance
class_counts = Counter([label for _, label in train_dataset.imgs])
class_sample_count = [class_counts[i] for i in range(num_classes)]
weights = 1. / torch.tensor(class_sample_count, dtype=torch.float)
samples_weight = [weights[label] for _, label in train_dataset.imgs]
sampler = WeightedRandomSampler(samples_weight, num_samples=len(samples_weight), replacement=True)

# Data loaders
train_loader = DataLoader(train_dataset, batch_size=batch_size, sampler=sampler, num_workers=2, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=2, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2, pin_memory=True)

# Initialize DenseNet169 (no pretrained weights - training from scratch)
model = models.densenet169(weights=None)
num_ftrs = model.classifier.in_features
model.classifier = nn.Linear(num_ftrs, num_classes)
model = model.to(device)

# Loss, optimizer, scheduler
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-2, weight_decay=1e-5)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=100, gamma=0.72)

# Initialize AMP GradScaler (do not specify device_type)
scaler = GradScaler()

def train_one_epoch(model, loader, criterion, optimizer):
    model.train()
    running_loss = 0.0
    running_corrects = 0
    for inputs, labels in tqdm(loader):
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        with autocast(device_type="cuda"):
            outputs = model(inputs)
            loss = criterion(outputs, labels)
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()
        running_loss += loss.item() * inputs.size(0)
        preds = outputs.argmax(dim=1)
        running_corrects += (preds == labels).sum().item()
        del inputs, labels, outputs, loss
        if device.type == 'cuda':
            torch.cuda.empty_cache()
    epoch_loss = running_loss / len(loader.dataset)
    epoch_acc = running_corrects / len(loader.dataset)
    return epoch_loss, epoch_acc

def evaluate(model, loader, criterion):
    model.eval()
    running_loss = 0.0
    running_corrects = 0
    with torch.no_grad():
        for inputs, labels in loader:
            inputs, labels = inputs.to(device), labels.to(device)
            with autocast(device_type="cuda"):
                outputs = model(inputs)
                loss = criterion(outputs, labels)
            running_loss += loss.item() * inputs.size(0)
            preds = outputs.argmax(dim=1)
            running_corrects += (preds == labels).sum().item()
            del inputs, labels, outputs, loss
            if device.type == 'cuda':
                torch.cuda.empty_cache()
    epoch_loss = running_loss / len(loader.dataset)
    epoch_acc = running_corrects / len(loader.dataset)
    return epoch_loss, epoch_acc

best_val_acc = 0
epochs_no_improve = 0

for epoch in range(num_epochs):
    train_loss, train_acc = train_one_epoch(model, train_loader, criterion, optimizer)
    val_loss, val_acc = evaluate(model, val_loader, criterion)
    scheduler.step()
    print(f"Epoch {epoch+1}/{num_epochs} - Train Loss: {train_loss:.4f} Train Acc: {train_acc:.4f} "
          f"Val Loss: {val_loss:.4f} Val Acc: {val_acc:.4f}")
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        epochs_no_improve = 0
        torch.save(model.state_dict(), os.path.join(base_dir, "best_model.pth"))
    else:
        epochs_no_improve += 1
        if epochs_no_improve >= patience:
            print("Early stopping triggered")
            break

# Load best model and test
model.load_state_dict(torch.load(os.path.join(base_dir, "best_model.pth")))
test_loss, test_acc = evaluate(model, test_loader, criterion)
print(f"Test Accuracy: {test_acc * 100:.2f}%")

# Inference with label mapping
idx_to_class = {v: k for k, v in train_dataset.class_to_idx.items()}
model.eval()
with torch.no_grad():
    for inputs, labels in test_loader:
        inputs = inputs.to(device)
        outputs = model(inputs)
        preds = outputs.argmax(dim=1)
        pred_labels = [idx_to_class[idx.item()] for idx in preds]
        true_labels = [idx_to_class[label.item()] for label in labels]
        for i in range(len(pred_labels)):
            print(f"Image {i+1}: Predicted: {pred_labels[i]}, Actual: {true_labels[i]}")
        break
```

```python
# pages/4_FetalMonitor.py
%%writefile pages/4_FetalMonitor.py
import streamlit as st
import torch
from torchvision import transforms, models
from PIL import Image
import torch.nn.functional as F
import pandas as pd
from fpdf import FPDF
import io
import random
import tempfile

# Page setup
st.set_page_config(page_title="Fetal Health Monitoring", layout="centered")

# Custom CSS for gradient background and styling
st.markdown("""
    <style>
        body {
            background: linear-gradient(to bottom, #fcb0b3 0%, #d8b4e2 50%, #add8e6 100%);
        }
        .stApp {
            background: linear-gradient(to bottom, #fcb0b3 0%, #d8b4e2 50%, #add8e6 100%);
            padding: 2rem;
        }
        .uploadedImage {
            border-radius: 12px;
            box-shadow: 0 0 10px rgba(0,0,0,0.1);
        }
    </style>
""", unsafe_allow_html=True)

st.title("🧠 Fetal Health Monitor - SmartMaternalCare")
st.markdown("Upload a fetal ultrasound image to check the health condition using our trained AI model.")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Class names matching training mapping: benign=0, malignant=1, normal=2
class_names = ['benign', 'malignant', 'normal']

# Load DenseNet169 model & weights
model = models.densenet169(pretrained=False)
num_ftrs = model.classifier.in_features
model.classifier = torch.nn.Linear(num_ftrs, len(class_names))
model_path = "/content/drive/MyDrive/mini_pro/fetal_densenet169_best.pth"
model.load_state_dict(torch.load(model_path, map_location=device))
model = model.to(device)
model.eval()

# Inference transforms for the image
inference_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225]),
])

# Medical comments and recommendations for each class
medical_comments = {
    'normal': [
        "The ultrasound scan shows typical fetal anatomy and structures with no abnormal patterns detected. Measurements such as head circumference, femur length, and amniotic fluid levels appear within normal ranges. This usually indicates healthy growth and development. Routine prenatal care should continue to monitor progress.",
        "Ultrasound indicates normal fetal growth and anatomy. No irregularities detected. Continue with routine prenatal checkups to ensure steady development.",
        "Fetal ultrasound findings are consistent with healthy development. All key parameters fall within expected ranges. Keep up regular monitoring for ongoing assurance."
    ],
    'benign': [
        "The scan revealed signs consistent with a benign condition — often non-threatening variations like small cysts, mild fluid accumulation, or temporary anatomical deviations. Follow up with your healthcare provider for proper monitoring and reassurance.",
        "Findings suggest a benign anomaly, which typically does not pose immediate risk. It is recommended to confirm with a follow-up ultrasound and consult your physician accordingly.",
        "Benign ultrasound features detected. These variations may resolve naturally but monitoring is necessary to ensure stability. Discuss results with your specialist."
    ],
    'malignant': [
        "This ultrasound shows patterns that may be associated with serious abnormalities, such as structural anomalies or growth restrictions. Urgent medical evaluation and specialist consultation are advised.",
        "Suspicious features detected in the scan that require prompt follow-up with advanced imaging and expert assessment to confirm diagnosis and planning.",
        "Abnormal ultrasound findings identified. Please seek immediate consultation with fetal medicine experts for comprehensive evaluation and management."
    ]
}

recommendations = {
    'normal': [
        "Continue your regular checkups and maintain a healthy lifestyle for the best outcomes.",
        "Your fetus appears healthy. Keep up the good work and stay positive.",
        "Everything looks normal. Regular monitoring is advised.",
        "You’re on the right track. Just continue following your doctor’s prenatal care plan.",
        "Stay hydrated, get enough rest, and attend all scheduled prenatal visits to ensure continued healthy development.",
        "No issues detected. Keep nurturing your pregnancy with love, care, and mindfulness.",
        "Things look great! Keep doing what you’re doing, and don’t forget your supplements and routine scans."
    ],
    'benign': [
        "Please schedule a follow-up appointment with your gynecologist to discuss these findings further.",
        "Consult your healthcare provider for reassurance and further tests if needed.",
        "Benign anomaly detected. Further monitoring is recommended.",
        "This is often not serious, but a second opinion from your healthcare provider is important.",
        "Keep calm. These are common and often resolve with time, but a follow-up scan is still necessary.",
        "You may be advised to undergo a targeted ultrasound to rule out any progression of the anomaly.",
        "Make a note of this finding and discuss it in your next prenatal appointment for tailored guidance."
    ],
    'malignant': [
        "Seek urgent medical evaluation to determine the best course of action and provide timely care.",
        "Immediate consultation with a specialist is highly recommended.",
        "Potential malignancy detected. Please contact your doctor as soon as possible.",
        "Follow-up imaging and specialist consultation should be arranged at the earliest opportunity.",
        "Time is crucial — please schedule an advanced diagnostic scan and consult a fetal medicine expert immediately.",
        "These signs could indicate a more serious issue. A team-based evaluation (OB-GYN, radiologist, neonatologist) may be needed.",
        "We strongly recommend a high-resolution scan and medical consultation within the next 24–48 hours."
    ]
}

affirmations = [
    "Dear mama, you are beautiful and strong. Trust yourself, be confident, and know your baby is waiting eagerly to meet you. Every step you take is filled with love and hope.",
    "You are radiant with life and courage. Believe in your strength and the beautiful journey ahead. Your baby feels your love and can’t wait to be in your arms soon.",
    "Mom, you are more powerful and amazing than you know. Stay calm, breathe deeply, and keep faith—your little one already loves you and is growing strong every day.",
    "Your body is a safe, warm home for your baby. Every breath you take nourishes life. You are doing everything right — just by being you.",
    "You are not alone. With every heartbeat, your baby feels your strength and your love. This bond you share is unbreakable and sacred.",
    "Each day you carry your child, you are writing a love story that will last a lifetime. You are enough. You are more than enough.",
    "Your love is already the strongest medicine your baby could ever receive. Trust the journey. You were made for this miracle.",
    "Even in moments of doubt, remember: your baby believes in you. You are already their whole world — safe, strong, and filled with love.",
    "Growing life is not easy, but you’re doing it beautifully. There is grace in your strength, and peace in your love.",
    "You are the calm in the storm, the light in the night. Your heartbeat is music to your baby’s soul.",
    "Every kick is a thank you. Every flutter is love. Your baby already knows your voice, your warmth, and your heart.",
    "This journey is uniquely yours — and every moment matters. You are becoming something divine: a mother.",
    "Let go of fear. Let in love. You are surrounded by life’s most sacred miracle. Be gentle with yourself today.",
    "Your softness is your superpower. Your patience, your tears, your courage — they are shaping the most precious life."
]

def create_medical_report(label, probs, comment, rec, pil_image):
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", "B", 16)

    pdf.cell(0, 10, "Fetal Health Monitoring Report", 0, 1, "C")
    pdf.ln(10)

    pdf.set_font("Arial", size=12)
    pdf.cell(0, 10, f"Prediction: {label.upper()}", 0, 1)
    pdf.ln(5)

    pdf.cell(0, 10, "Confidence Scores:", 0, 1)
    for cls, prob in zip(class_names, probs):
        pdf.cell(0, 10, f"{cls.capitalize()}: {prob*100:.2f} %", 0, 1)
    pdf.ln(5)

    pdf.multi_cell(0, 10, f"Medical Comment: {comment}")
    pdf.ln(5)

    pdf.multi_cell(0, 10, f"Recommendation: {rec}")
    pdf.ln(10)

    # Embed PIL image temporarily
    with tempfile.NamedTemporaryFile(suffix=".png") as tmp_file:
        pil_image.save(tmp_file.name, format="PNG")
        pdf.image(tmp_file.name, x=10, w=pdf.w - 20)  # leave 10 margin on both sides

    pdf_str = pdf.output(dest='S')  # returns string
    pdf_bytes = pdf_str.encode('latin1')  # encode string to bytes for BytesIO
    pdf_output = io.BytesIO(pdf_bytes)
    pdf_output.seek(0)
    return pdf_output


# Upload fetal ultrasound image
uploaded_file = st.file_uploader("📤 Upload a fetal ultrasound image", type=["jpg", "jpeg", "png"])

if uploaded_file:
    image = Image.open(uploaded_file).convert("RGB")
    input_tensor = inference_transform(image).unsqueeze(0).to(device)

    with torch.no_grad():
        output = model(input_tensor)
        probs = F.softmax(output, dim=1).cpu().numpy()[0]
        pred_class_idx = probs.argmax()
        label = class_names[pred_class_idx]

    # Random medical comment and recommendation
    comment = random.choice(medical_comments[label])
    rec = random.choice(recommendations[label])

    # Display prediction
    st.markdown(f"### 🔍 Prediction: **{label.upper()}**")

    # Confidence bar chart
    df_probs = pd.DataFrame({'Class': class_names, 'Probability': probs})
    df_probs = df_probs.sort_values('Probability', ascending=False).reset_index(drop=True)
    st.bar_chart(df_probs.set_index('Class'))

    # Show comment and recommendation with style
    if label == 'normal':
        st.success(comment)
        st.info(rec)
    elif label == 'benign':
        st.warning(comment)
        st.info(rec)
    else:
        st.error(comment)
        st.warning(rec)

    # Show uploaded image
    st.image(image, caption="🖼 Uploaded Fetal Ultrasound Image", use_container_width=True)

    # Generate downloadable PDF report with embedded image
    report_pdf = create_medical_report(label, probs, comment, rec, image)
    st.download_button(
        label="📄 Download Medical Report (PDF)",
        data=report_pdf,
        file_name="fetal_health_report.pdf",
        mime="application/pdf"
    )

    # Uplifting affirmation
    st.markdown("---")
    st.subheader(f"💖 {random.choice(affirmations)}")
```

```python
#chatbot
%%writefile pages/3_BellyBot.py
import streamlit as st
import google.generativeai as genai
import time

# ===============
# CONFIGURATION
# ===============
USER_ICON = "🤰"
BOT_ICON = "🤖"
API_KEY = "AIzaSyADoiMayhQlbeBkSw9GOmSTPItNIzlMr5E"

genai.configure(api_key=API_KEY)

MODEL_NAME = "gemini-2.5-flash"
INSTRUCTION = """
Pregnancy Assistant.
Always provide supportive, friendly responses (3–5 sentences),
including possible causes, helpful daily measures, and when to consult a doctor.
Be empathetic, calm, and clear.
Only include a short reference link to a reliable health source
(such as WHO, Mayo Clinic, or ACOG) if the user specifically requests references or sources.
Always remind the user that your advice does not replace professional medical guidance,
and encourage consulting a healthcare provider for any serious or persistent concerns.
"""

# ==============
# STREAMLIT UI
# ==============
st.set_page_config(page_title="Pregnancy Chatbot", layout="centered")
st.markdown("""
<h1 style='text-align: center; color: #495057;'>
    🤖 PregoCare
</h1>
""", unsafe_allow_html=True)

# Welcome message
st.markdown("""
<div style="background: rgba(255,255,255,0.88); border-radius: 16px; padding: 16px; margin-bottom: 16px;">
<h3 style="color: #b264a6; text-align: center; font-weight: bold; margin-top:0;">
Your Pregnancy Companion!
</h3>
<p style="font-size: 18px; color: #495057; text-align: center;">
Chat with your personal assistant about health, mood, or daily tips during pregnancy. 💖
</p>
</div>
""", unsafe_allow_html=True)

# ===============
# SESSION STATE
# ===============
if "chat_history" not in st.session_state:
    st.session_state.chat_history = []

if "chat_session" not in st.session_state:
    model = genai.GenerativeModel(
        MODEL_NAME,
        system_instruction=INSTRUCTION
    )
    st.session_state.chat_session = model.start_chat(history=[])

# ======================
# GET RESPONSE FUNCTION
# ======================
def get_response(query: str) -> str:
    try:
        response = st.session_state.chat_session.send_message(query)
        return response.text if hasattr(response, 'text') and response.text else \
            "🤖 Sorry, I received an empty response from the AI."
    except Exception as e:
        return f"🤖 An error occurred: {e}. I'm here for you."

# ==========================
# TYPEWRITER EFFECT GENERATOR
# ==========================
def stream_generator(text: str):
    for chunk in text.split():
        yield chunk + " "
        time.sleep(0.05)

# =====================
# DISPLAY CHAT HISTORY
# =====================
for i, item in enumerate(st.session_state.chat_history):
    role, msg, is_streamed = item

    if role == "You":
        with st.chat_message("user", avatar=USER_ICON):
            st.markdown(msg)

    elif role == "Bot":
        if not is_streamed:
            with st.chat_message("assistant", avatar=BOT_ICON):
                st.write_stream(stream_generator(msg))
            st.session_state.chat_history[i] = (role, msg, True)
            st.rerun()
        else:
            with st.chat_message("assistant", avatar=BOT_ICON):
                st.markdown(msg)

# ==========================
# LOADER PLACEHOLDER (above textbox)
# ==========================
loader_placeholder = st.empty()

# ==========================
# CHAT INPUT FORM (sticky)
# ==========================
st.markdown(
    """
    <style>
    .stApp {
      background: linear-gradient(to bottom, #fcb0b3 0%, #d8b4e2 50%, #add8e6 100%);
      color: #000000;
    }
    .chat-input-container {
        position: fixed;
        bottom: 0;
        left: 0;
        right: 0;
        z-index: 100;
    }
    .stTextInput>div>div>input {
        padding: 10px;
        font-size: 16px;
    }
    </style>
    """,
    unsafe_allow_html=True
)

st.markdown("<div style='height:90px;'></div>", unsafe_allow_html=True)

with st.container():
    st.markdown("<div class='chat-input-container'>", unsafe_allow_html=True)

    with st.form(key='chat_form', clear_on_submit=True):
        user_input = st.text_input(
            "You:",
            key='chat_input_key',
            placeholder="Ask about symptoms, diet, or tips...",
            label_visibility="collapsed"
        )
        submitted = st.form_submit_button("Send")

    st.markdown("</div>", unsafe_allow_html=True)

# ==========================
# SEND LOGIC
# ==========================
if submitted and st.session_state.chat_input_key:
    query = st.session_state.chat_input_key
    st.session_state.chat_history.append(("You", query, True))

    with loader_placeholder.container():
        with st.spinner("**PregoCare is thinking✨...**"):
            bot_reply = get_response(query)

    st.session_state.chat_history.append(("Bot", bot_reply, False))
    st.rerun()
```
